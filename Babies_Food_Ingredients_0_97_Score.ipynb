```python
import numpy as np
import pandas as pd
from sklearn.impute import KNNImputer
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestRegressor
import xgboost as xg
```

```python
df=pd.read_csv('food_Ingredients.csv')
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.head(10)
```

```python
df['size']=df['size'].fillna(df['size'].mean())
```

```python
df.corr()
```

```python
my_pipeline=Pipeline(steps=[('model',xg.XGBRegressor())])
```

```python
X=df.drop(['quality'],axis=1)
y=df['quality']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
my_pipeline.fit(X_train,y_train)
```

```python
my_pipeline.score(X_test,y_test)
```

```python
X=df.drop(['quality'],axis=1)
y=df['quality']
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=10)
from sklearn.feature_selection import SelectKBest,f_classif
np.seterr(divide='ignore',invalid='ignore')
t=SelectKBest(f_classif,k=15).fit(X_train,y_train)
st=X.columns[t.get_support()].tolist()
print(st)
```

```python
feat = ['size', 'calories_kcal', 'carb_g', 'fiber_g', 'sugar_g', 'protein_g', 'potassium_mg', 'magnesium_mg', 'vitE_mg', 'vitC_mg', 'vitB6_mg', 'copper_mg', 'carotene_mg', 'cryptoxanthin_mcg', 'lycopene_mcg']
X=df[feat]
y=df['quality']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
my_pipeline.fit(X_train,y_train)
my_pipeline.score(X_test,y_test)
```

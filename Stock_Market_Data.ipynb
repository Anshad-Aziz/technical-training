```python
import yfinance as yf
```

```python
datasets=[
     "^NSEBANK",
    "ICICIBANK.NS",
    "HDFCBANK.NS",
    "FEDERALBNK.NS",
    "SBIN.NS",
    "AUBANK.NS",
    "BANKBARODA.NS",
    "INDUSINDBK.NS",
    "PNB.NS",
    "AXISBANK.NS",
    "IDFCFIRSTB.NS",
    "KOTAKBANK.NS",
    "BANDHANBNK.NS"
]

for dataset in datasets:
  ticker=yf.Ticker(dataset)
  data=ticker.history(start="2012-01-01",end="2022-12-31")
  filename=f"{dataset}_data.csv"
  data.to_csv(filename)
  print(f"Dowloaded data for {dataset} and saved as {filename}.")
```

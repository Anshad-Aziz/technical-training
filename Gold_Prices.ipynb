```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('gold-prices.csv',index_col='Date',parse_dates=True)
df
```

```python
df.tail()
```

```python
plt.figure(figsize=(16,7))
#sns.lineplpot(data=df)
x=plt.xticks(rotation=90)
sns.lineplot(data=df['Price'],color='r')
plt.grid(True)
```

```python
plt.figure(figsize=(16,7))
sns.histplot(df['Price'],color='r')
plt.grid(True)
```

```python
plt.figure(figsize=(16,7))
sns.kdeplot(df['Price'],fill=True,color='r')
plt.grid(True)
```
